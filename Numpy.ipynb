4)  **Array Function**

```python
c1 = np.random.random((3,4))
c1 = np.round(c1 * 100)
c1
```

```python
# Min
print("Min :",np.min(c1))

# Max
print("Max :",np.max(c1))

# Sum
print("Sum :",np.sum(c1))

# product
print("product :",np.product(c1))
```

```python
# 0 --> Col and 1 --> Row
print("Min in every col :", np.min(c1, axis=0))
print("Min in every row :", np.min(c1, axis=1))
print("product of every col :",np.product(c1, axis=0))
```

```python
# Mean /Median /std /var
```

```python
# Dot product
d1 = np.arange(12).reshape(3,4)
d2 = np.arange(12,24).reshape(4,3)
print(d1,'\n\n',d2)

# matrix 1 col and matrix 2 row must be equal

print('\n',np.dot(d1,d2))
```

7) **Iterating**

```python
print(z1,'\n')

for i in z1:
  print(i)
```

```python
print(z2,'\n')

for i in z2:
  print(i,'\n') # In 2D a row print in 1 time
```

```python
print(z3,'\n')

for i in z3:
  print(i,'\n') # In 3D a 2D printed in 1 time
```

```python
# To print every single item in 2D,3D etc we use nditer(convert array in 1D than print)
for i in np.nditer(z3):
  print(i)
```

9) **Stacking**

```python
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
print(a4,'\n')
print(a5)
```

```python
# Horizontal stacking
np.hstack((a4,a5)) # shape must be same
```

```python
# Vertical stacking
np.vstack((a4,a5))
```
